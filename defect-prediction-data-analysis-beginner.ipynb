```python
import matplotlib.pyplot as plt    
import seaborn as sns               
import plotly       
import pandas as pd        
import numpy as np 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from lightgbm import LGBMClassifier
```

```python
train = pd.read_csv("/kaggle/input/playground-series-s4e3/train.csv")
test = pd.read_csv("/kaggle/input/playground-series-s4e3/test.csv")
```

```python
target_labels = ["Pastry","Z_Scratch","K_Scatch","Stains","Dirtiness","Bumps","Other_Faults"]
X_train = train.drop(target_labels, axis=1)
y_train = train[target_labels]

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
model = LGBMClassifier(n_estimators=200, verbose=-1, random_state=42, min_child_weight=0.01, reg_alpha=0.2)
multi_output_model = MultiOutputClassifier(model)
multi_output_model.fit(X_train, y_train)
```

```python
accuracy = multi_output_model.score(X_val, y_val)
print("Validation Accuracy:", accuracy)
```

```python
test_predictions = multi_output_model.predict_proba(test)
positive_class_probabilities = np.array([test_predictions[i][:,1] for i in range(len(test_predictions))])
```

```python
submission = pd.DataFrame(positive_class_probabilities.T, columns=target_labels)
submission['id'] = test['id']
submission.to_csv('submission.csv', index=False)
```
